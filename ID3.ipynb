```python
import seaborn as sns
```

```python
data=sns.load_dataset('iris')
x = data.drop("species", axis=1)
y = data["species"]
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
y=encoder.fit_transform(y)
```

```python
#Generate test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#create the decision tree

from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier(criterion="entropy",max_depth=3)
dtree.fit(x_train,y_train)
print("Decision tree classifier created")
y_pred=dtree.predict(x_test)
```

```python
from sklearn import metrics
print("ID3 model accuracy : ",metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(10,10))
dec_tree=plot_tree(decision_tree=dtree,feature_names=data.columns,class_names=["setosa","vercicolor","verginica"],
                   filled=True,precision=4,rounded=True)
```
