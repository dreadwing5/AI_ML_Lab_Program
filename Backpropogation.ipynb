```python
import numpy as np
```

```python
input_layer_neurons=2
hidden_layer_neurons=3
output_neurons=1
print("Number of input layer neurons: ",input_layer_neurons)
print("Number of hidden layer neurons: ",hidden_layer_neurons)
print("Number of output layer neurons: ",output_neurons)
```

```python
wh=np.random.uniform(size=(input_layer_neurons,hidden_layer_neurons))
print("Weights from input layer to hidden layer : \n",wh)
bh=np.random.uniform(size=(1,hidden_layer_neurons))
print("Bias between input and hidden layers : \n",bh)
wout=np.random.uniform(size=(hidden_layer_neurons,output_neurons))
print("Weights from hidden layer to output layer : \n",wout)
bout=np.random.uniform(size=(1,output_neurons))
print("Bias between output and hidden layers : \n",bout)
```

```python
epoch=7000
lr=0.1
x=np.array(([2,9],[1,5],[3,6]))
y=np.array(([92],[86],[89]))
x=x/np.amax(x,axis=0)
x
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def derivatives_sigmoid(x):
    return x*(1-x)
```

```python
for i in range(epoch):
    hinp1=np.dot(x,wh)
    hinp=hinp1+bh
    hlayer_act=sigmoid(hinp)
    outinp1=np.dot(hlayer_act,wout)
    outinp=outinp1+bout
    output=sigmoid(outinp)
    EO=y-output
    outgrad=derivatives_sigmoid(output)
    d_output=EO*outgrad
    EH=d_output.dot(wout.T)
    hiddengrad=derivatives_sigmoid(hlayer_act)
    d_hiddenlayer=EH*hiddengrad
    wout+=hlayer_act.T.dot(d_output)*lr
    wht=x.T.dot(d_hiddenlayer)*lr
    print("Input : \n",str(x))
    print("Actual o/p : \n",str(y))
    print("Predicted o/p : \n",output)
```
